```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.ticker as ticker
import IPython.display

plt.rcParams['font.family'] = ["D2Coding"]

from mendeleev import element
```

# 원자의 주기성 시각화

`mendeleev` 사용해봄

* 원자 반지름
* 이온화 에너지
* 전기음성도

```python
domain = list(range(1, 20+1))

e18 = [2, 10, 18]

fig, ax = plt.subplots(1, 3, figsize=(21, 6))
ax[0].bar(*zip(*[(e, element(e).atomic_radius) for e in domain if e not in e18]), color='red')
ax[0].set_title("원자 반지름")
ax[0].set_xticks(domain)
ax[1].bar(domain, [element(e).ionenergies[1] for e in domain], color='blue')
ax[1].set_title("이온화 에너지 $E_1$")
ax[1].set_xticks(domain)
ax[2].bar(*zip(*[(e, element(e).electronegativity()) for e in domain if e not in e18]), color='green')
ax[2].set_title("전기음성도")
ax[2].set_xticks(domain)
plt.show()
```

```python
domain = list(range(1, 20+1))

domain = [7, 8, 9, 10, 11, 12, 13]
domain2 = [16, 17, 18, 19, 20]

def f(ionic_radii, want):
    return list(filter(lambda x: x.charge == want, ionic_radii))[0]

e18 = [2, 10, 18]
fig, ax = plt.subplots(1, 2, figsize=(20, 7))
ax[0].plot(*zip(*[(e, f(element(e).ionic_radii, e-10).ionic_radius) for e in domain if e not in e18]), color='blue')
ax[0].plot(*zip(*[(e, element(e).atomic_radius) for e in domain if e not in e18]), color='red')
ax[0].set_title("등전자 이온 반지름(Ne-like)")
ax[0].set_xticks(domain)
ax[0].legend(["이온 반지름", "원자 반지름"])

ax[1].plot(*zip(*[(e, f(element(e).ionic_radii, e-18).ionic_radius) for e in domain2 if e not in e18]), color='blue')
ax[1].plot(*zip(*[(e, element(e).atomic_radius) for e in domain2 if e not in e18]), color='red')
ax[1].set_title("등전자 이온 반지름(Ar-like)")
ax[1].set_xticks(domain2)
ax[1].legend(["이온 반지름", "원자 반지름"])
plt.show()
```
