```python
import csv
import datetime

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.ticker as ticker
import IPython.display


plt.rcParams['font.family'] = ["D2Coding"]
```

```python
with open("dataset/birth/한국_출산통계_2020.csv", newline='', encoding='utf-8') as f:
    _csv = csv.DictReader(f)
    for i, row in enumerate(_csv):
        if i == 4:
            data = row

TFR = dict()
years = list(range(1970, 2020+1))
for year in years:
    TFR[year] = float(data[str(year)])
    print(year, TFR[year])

```

```python
fig, ax = plt.subplots( figsize=(8, 4))

ax.set_ylim((0, 5))
ax.plot(years, [TFR[y] for y in years], color="darkslateblue", label="합계출산율")
ax.set_ylabel("합계 출산율")

ax2 = ax.twinx()
ax2.set_ylim((-0.5, 0.5))
ax2.plot([ys for yf, ys in zip(years[:-1], years[1:])],
        [TFR[ys]-TFR[yf] for yf, ys in zip(years[:-1], years[1:])], color="crimson", label="합계출산율 변화")
ax2.set_title("합계출산율 실태와 변화")
ax2.set_ylabel("변화")
ax.set_xlabel("연도")

fig.legend(loc='upper right', bbox_to_anchor=(1, 1), bbox_transform=ax.transAxes)

plt.show()
```
