# 수능 단어 분석


```python
with open("dataset/english_word_frequency/englsih_texts/data.pkl", 'rb') as f:
    data = pickle.load(f)

q = {}

for word in data:
    if word in q:
        q[word] += 1
    else:
        q[word] = 1

print(max(q.items(), key=lambda x: x[1]))
print(sorted(q.items(), key=lambda x: x[1])[::-1])
# print(sorted(q.items(), key=lambda x: x[1])[::-1])
```
