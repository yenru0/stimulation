```python
import pickle

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.ticker
import IPython.display

plt.rcParams["font.family"] = "D2Coding"
```

# 수능 단어 분석


```python
with open("dataset/english_word_frequency/englsih_texts/data.pkl", 'rb') as f:
    data = pickle.load(f)

q = {}

for word in data:
    if word in q:
        q[word] += 1
    else:
        q[word] = 1

words = list(q.items())
words_sorted = np.array(sorted(q.items(), key=lambda x: x[1])[::-1])
```

```python
fig, ax = plt.subplots(figsize=(20, 7))

x = words_sorted[:36, 0]
y = words_sorted[:36, 1].astype(int)
ax.bar(x, y, color=plt.cm.coolwarm(mpl.colors.LogNorm (1, np.max(y))(y)))

ax.set_title("수능 빈출 단어")

plt.show()
```
