# 포식자-피식자 관계
**Lotka-Volterra 방정식**은 포식자 피식자 관계를 수량화한 연립 미분 방정식이다.

이 방정식은 다음과 같다.

$$\frac{dx}{dt} = \alpha x - \beta xy$$

$$\frac{dy}{dt} = \delta xy - \gamma y$$

```python
fig, ax = plt.subplots(figsize=(10, 6))

t, dt = np.linspace(0, 30, 100000, retstep= True)

alpha = 1
beta = 0.5
gamma = 0.5
delta = 0.25

# 초깃값
prey     = [5,]
predator = [0.04]

for i in range(len(t) - 1):
    new_prey = prey[-1] + dt * (prey[-1] * alpha - beta * prey[-1] * predator[-1])
    new_predator = predator[-1] + dt * (-predator[-1] * gamma + delta * prey[-1] * predator[-1])
    prey.append(new_prey)
    predator.append(new_predator)

ax.plot(t, prey, color='green', label='피식자')
ax.plot(t, predator, color='red', label='포식자')

ax.grid()

ax.legend()

plt.show()
```
