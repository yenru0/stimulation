```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open("./resources/partialize.png")


plt.imshow(image)
plt.show()
```

```python
def make_partial(img, size = (256, 256), partials = (2, 2)):
    dx, dy = size[0] // partials[0], size[1] // partials[1]
    data = np.array(img)
    im = np.zeros((partials[0], partials[1], 4), dtype=np.uint8)
    for i in range(partials[0]):
        for j in range(partials[1]):
            t = np.array((np.mean(data[i*dx: dx*(i+1), j*dy: dy*(j+1)], axis=(0, 1))), dtype=np.uint8)
            im[i, j] = t
    return im
```

```python
fig, ax = plt.subplots(4, 2, figsize=(5, 10))

ax[0, 0].imshow(make_partial(image, size=(256, 256), partials=(1, 1)))
ax[0, 1].imshow(make_partial(image, size=(256, 256), partials=(2, 2)))
ax[1, 0].imshow(make_partial(image, size=(256, 256), partials=(4, 4)))
ax[1, 1].imshow(make_partial(image, size=(256, 256), partials=(8, 8)))
ax[2, 0].imshow(make_partial(image, size=(256, 256), partials=(16, 16)))
ax[2, 1].imshow(make_partial(image, size=(256, 256), partials=(32, 32)))
ax[3, 0].imshow(make_partial(image, size=(256, 256), partials=(64, 64)))
ax[3, 1].imshow(make_partial(image, size=(256, 256), partials=(128, 128)))

fig.show()
```
