# 극함수 보간

$r=f(\theta)$를 만들어 임의의 점 $(r_i, \theta_i)$에 대해서 $f(\theta_i)=r_i$를 만족하는 함수 $f$를 만들면 보간될 수 있다고 생각한다.

일반적인 보간을 조금만 변형하면 될 것 같다.

```python
def cardinal(xset, i, x):
    n = len(xset)
    return np.prod([(x-xset[j])/(xset[i] - xset[j]) for j in range(n) if i != j])

def lagrangian_form(xset, yset, x):
    n = len(xset)
    return np.sum([cardinal(xset, i, x)*yset[i] for i in range(n)])

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

_rs  = [-1, 2, 6, 8]
_ths = [3, 2, 1, 0]

thetas = np.linspace(0, 6*np.pi, 1000)

R = [lagrangian_form(_rs, _ths, theta) for theta in thetas]

X = [r*np.cos(theta) for r, theta in zip(R, thetas)]
Y = [r*np.sin(theta) for r, theta in zip(R, thetas)]


ax[0].grid()
ax[0].set_xlim((-3, 3))
ax[0].set_ylim((-3, 3))
ax[0].plot(X, Y, color='red')

ax[1].grid()
ax[1].set_xlim((0, 6*np.pi))
ax[1].set_ylim((-5, 5))
ax[1].plot(thetas, R, color='blue')

plt.show()
```
