# 바이어슈트라스 함수(Weierstrass function)

병적인(Pathological) 함수의 한 예이며 **모든 점에서 연속이나 모든 점에서 미분 불가능한 함수**의 대표적인 예, 카를 바이어슈트라스가 제안하였다.

바이어 슈트라스 함수의 논문 원본에서의 정의는 다음과 같다.

$$f(x)=\sum^{\infty}_{n=0}{a^n \cos(b^n\pi x)}$$

(단, $0 < a < 1,\, b는 양의 홀수,\, ab>1+\frac{3\pi}{2}$)

하지만 $0 < a < 1,\, ab\geq1$이라는 전제 하에 모든 점에서 미분 불능임을 하디가 증명했다.


```python
plt.figure(figsize=(20, 10))

def weierstrass(domain, a, b, n = 100):
    return np.sum([a**i * np.cos(b**i*np.pi*domain) for i in range(n+1)], axis=0)

domain = np.linspace(-2, 2, 100000)

plt.plot(domain, weierstrass(domain, 0.5, 3))

plt.show()
```
