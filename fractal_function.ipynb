# 바이어슈트라스 함수(Weierstrass Function)

병적인(Pathological) 함수의 한 예이며 **모든 점에서 연속이나 모든 점에서 미분 불가능한 함수**의 대표적인 예, 카를 바이어슈트라스가 제안하였다.

바이어 슈트라스 함수의 논문 원본에서의 정의는 다음과 같다.

$$f(x)=\sum^{\infty}_{n=0}{a^n \cos(b^n\pi x)}$$

(단, $0 < a < 1,\, b는 양의 홀수,\, ab>1+\frac{3\pi}{2}$)

하지만 $0 < a < 1,\, ab\geq1$이라는 전제 하에 모든 점에서 미분 불가능임을 하디가 증명했다.

```python
def weierstrass(a, b, n=100):
    return lambda x: np.sum([a ** i * np.cos(b ** i * np.pi * x) for i in range(n + 1)], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

x = np.linspace(-2, 2, 10000)

w = weierstrass(0.5, 3)

ax.plot(x, w(x), color='red')

ax.set_title("Weierstrass Function")

plt.show()
```

# 망델브로 집합(Mandelbrot Set)

망델브로 집합은 다음 점화식으로 정의된 수열 $z_n$이 발산하지 않도록 하는 복소수 $c$의 집합으로 정의한다.

$$\begin{align}
&z_0 = 0 \\
&z_{n+1} = z_0^2 + c
\end{align}$$

만약 복소수를 사용하지 않기 위해선 다음과 같이 $x_n$과 $y_n$을 나눠 정의하면 된다.

$$\begin{align}
&(x_0,y_0) = (0,0)\\
&x_{n+1} = x_n^2 - y_n^2 + a \\
&y_{n+1} = 2x_n y_n + b
\end{align}$$

```python
#a, b = np.linspace(-0.9, -0.85, 1000), np.linspace(0.23, 0.28, 1000)
a, b = np.linspace(-2, 1, 1000), np.linspace(-1.5, 1.5, 1000)
A, B = np.meshgrid(a, b)

x, y = np.zeros_like(a), np.zeros_like(b)
X, Y = np.meshgrid(x, y)
C = np.zeros_like(X)

for i in range(48):
    X_new = (X**2 - Y**2) + A
    Y_new = (2*X*Y) + B
    X = X_new
    Y = Y_new
    C[np.logical_and(X**2+Y**2 > 4, C==0)] = i+1
    X[X > 10] = 10
    Y[Y > 10] = 10
    
    
S = X**2 + Y**2 < 4
A_m = A[S]
B_m = B[S]
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(A_m, B_m, s=2, color='black')
plt.show()
```
